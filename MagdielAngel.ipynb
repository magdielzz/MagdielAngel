1. **¿Qué significa la palabra "Flop"? ¿Qué es Rmax y Rpeak?**

   - **Flop**: Es la abreviatura de "Floating Point Operation" (Operación de Punto Flotante). Se utiliza para medir el rendimiento de las computadoras, especialmente en cálculos científicos que requieren operaciones matemáticas complejas. La capacidad de una computadora se mide en FLOPS (Floating Point Operations Per Second), indicando cuántas operaciones de punto flotante puede realizar por segundo.

   - **Rpeak**: Es el rendimiento teórico máximo que una supercomputadora puede alcanzar. Se calcula en función de las especificaciones del hardware, como la frecuencia del procesador y el número de unidades de procesamiento. Representa el límite superior de rendimiento bajo condiciones ideales.

   - **Rmax**: Es el rendimiento máximo real medido de una supercomputadora al ejecutar el benchmark LINPACK, que resuelve sistemas lineales de ecuaciones. Este valor suele ser inferior a Rpeak debido a las ineficiencias prácticas en la ejecución de tareas.

### 9. ¿Qué es la tabla de procesos en Linux? ¿Qué es un proceso?

**Proceso:**
En Linux, un **proceso** es una instancia de un programa en ejecución. Cada proceso tiene su propio espacio de memoria, recursos asignados (como archivos abiertos, conexiones de red, etc.) y un estado de ejecución. Los procesos son gestionados por el kernel del sistema operativo, que se encarga de su creación, planificación, ejecución y terminación.

**Tabla de procesos:**
La **tabla de procesos** es una estructura de datos mantenida por el kernel de Linux que almacena información sobre todos los procesos activos en el sistema. Cada entrada en la tabla de procesos corresponde a un proceso y contiene detalles como:

- **PID (Process ID):** Identificador único del proceso.
- **Estado del proceso:** Por ejemplo, en ejecución, en espera, detenido, etc.
- **Prioridad:** Información sobre la prioridad de planificación del proceso.
- **Registros de la CPU:** Estado de los registros del procesador cuando el proceso no está en ejecución.
- **Espacio de memoria:** Información sobre la memoria asignada al proceso.
- **Propietario:** El usuario y grupo que poseen el proceso.
- **Recursos abiertos:** Archivos, sockets, etc., que el proceso tiene abiertos.

La tabla de procesos es esencial para que el kernel pueda gestionar y planificar la ejecución de múltiples procesos de manera 10ficiente.

### 2. ¿Cuál es la diferencia entre un hilo (thread) y un proceso?

**Proceso:**
- Un **proceso** es una entidad independiente que tiene su propio espacio de memoria, recursos y estado de ejecución.
- Los procesos no comparten memoria entre sí (a menos que se utilicen mecanismos específicos como memoria compartida).
- La creación de un nuevo proceso (mediante `fork()` en Linux) es más costosa en términos de recursos, ya que implica duplicar el espacio de memoria y otros recursos.
- Los procesos se comunican entre sí mediante mecanismos de IPC (Inter-Process Communication), como tuberías (pipes), colas de mensajes, memoria compartida, etc.

**Hilo (Thread):**
- Un **hilo** es una unidad de ejecución dentro de un proceso. Todos los hilos de un proceso comparten el mismo espacio de memoria y recursos.
- Los hilos son más ligeros que los procesos, ya que no requieren duplicar el espacio de memoria. Esto hace que la creación y el cambio de contexto entre hilos sea más rápido.
- Los hilos pueden comunicarse entre sí directamente a través de la memoria compartida, lo que facilita la comunicación y la sincronización.
- Sin embargo, debido a que comparten memoria, los hilos deben sincronizarse cuidadosamente para evitar condiciones de carrera (race conditions) y otrodentro de un mismo espacio de memoria.
